```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
iris["data"] = np.asarray(iris["data"])
iris["attribute_names"] = np.array(['sepal length','sepal width','petal length','petal width'])
```

```python
fig, ax = plt.subplots(3, 4, sharex='col', figsize=(20,10))

colors = ['r', 'g', 'b']

for i in range(4):
    for j in range(3):
        ax[j, i].hist(np.split(iris["data"][:,i], 3)[j], 50, color=colors[j])
        ax[j, i].set_title(iris["attribute_names"][i] + " " + iris["target_names"][j])
        ax[j, i].tick_params('both', reset=True)

plt.suptitle("Histograms")
plt.show()
```
